```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = df[['TV']]  # Independent variable (TV budget)
y = df['sales']  # Dependent variable (Sales)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

# Get the slope (coefficient) and intercept
print("Slope: ",{model.coef_[0]})
print("Intercept: ",{model.intercept_})
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=X_test['TV'], y=y_test, color='blue', label="Actual Sales")
sns.lineplot(x=X_test['TV'], y=y_pred, color='red', label="Predicted Sales")
plt.xlabel("TV Budget ($)")
plt.ylabel("Sales")
plt.title("Simple Linear Regression (TV vs Sales)")
plt.legend()
plt.show()
```
